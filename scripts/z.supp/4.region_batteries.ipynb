```python
import MultiTaskBattery.battery as bat
import Functional_Fusion.atlas_map as am
import numpy as np
```

```python
base_dir='Y:/data/FunctionalFusion_new'
```

# Cortical batteries

```python
atlas = 'fs32k'
atlas_obj,_= am.get_atlas(atlas)
data_cortex,info = bat.fetch_task_library(atlas= atlas,version='V1')
full_code = info['task_code'] + '_' + info['cond_code']
```

Motor and PFC ROIS

```python
atlas_dir = f'{base_dir}/Atlases/tpl-fs32k'
model_name_L = f'{atlas_dir}/glasser.L.label.gii'
model_name_R = f'{atlas_dir}/glasser.R.label.gii'
parcels = atlas_obj.read_data([model_name_L,model_name_R])

ROI_motor = np.isin(parcels, [53,9,8,39,36,37,96,54,44,55,78,56,42]).astype(int)[np.newaxis, :]
ROI_PFC = np.isin(parcels, [74,81,82,83,84,86]).astype(int)[np.newaxis, :]
```

Motor and PFC batteries


```python
data_motor = data_cortex * ROI_motor
motor_df= bat.get_top_batteries(data_motor,full_code,n_samples=1000000, battery_size=8, n_top_batteries=10,forced_tasks=None)
display(motor_df)
```

```python
motor_df['tasks'][0]
```

```python
data_pfc = data_cortex * ROI_PFC
pfc_df= bat.get_top_batteries(data_pfc,full_code,n_samples=1000000, battery_size=8, n_top_batteries=10,forced_tasks=None)
display(pfc_df)
```
