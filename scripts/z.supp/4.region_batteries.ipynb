```python
import MultiTaskBattery.battery as bat
import Functional_Fusion.atlas_map as am
import numpy as np
from OptimalBattery.global_config import data_dir
```

# Cortical batteries

```python
atlas = 'fs32k'
atlas_obj,_= am.get_atlas(atlas)
data_cortex,info = bat.fetch_task_library(atlas= atlas,version='V1')
```

Motor and PFC ROIS

```python
atlas_dir = f'{data_dir}/FunctionalFusion_new/Atlases/tpl-fs32k'
model_name_L = f'{atlas_dir}/glasser.L.label.gii'
model_name_R = f'{atlas_dir}/glasser.R.label.gii'
parcels = atlas_obj.read_data([model_name_L,model_name_R])

ROI_motor = np.isin(parcels, [53,9,8,39,36,37,96,54,44,55,78,56,42]).astype(int)[np.newaxis, :]
ROI_PFC = np.isin(parcels, [74,81,82,83,84,86]).astype(int)[np.newaxis, :]
```

Motor and PFC batteries


```python
data_motor = data_cortex * ROI_motor
motor_df_4= bat.get_top_batteries(data_motor,info,n_samples=1000000, battery_size=4, n_top_batteries=5,forced_tasks=['rest_task'], verbose= False)
print('size 4 motor battery:')
print(motor_df_4['battery'][0])
motor_df_6= bat.get_top_batteries(data_motor,info,n_samples=1000000, battery_size=6, n_top_batteries=5,forced_tasks=['rest_task'], verbose= False)
print('size 6 motor battery:')
print(motor_df_6['battery'][0])
motor_df_8= bat.get_top_batteries(data_motor,info,n_samples=1000000, battery_size=8, n_top_batteries=5,forced_tasks=['rest_task'], verbose= False)
print('size 8 motor battery:')
print(motor_df_8['battery'][0])
```

```python
motor_df_6= bat.get_top_batteries(data_motor,info,n_samples=1000000, battery_size=6, n_top_batteries=100,forced_tasks=['rest_task'], verbose= False)
```

```python
task_counts = motor_df_6['battery'].explode().value_counts()
task_counts
```

```python
data_pfc = data_cortex * ROI_PFC
pfc_df_4= bat.get_top_batteries(data_pfc,info,n_samples=1000000, battery_size=4, n_top_batteries=5,forced_tasks=['rest_task'],verbose= False)
print('size 4 pfc battery:')
print(pfc_df_4['battery'][0])
pfc_df_6= bat.get_top_batteries(data_pfc,info,n_samples=1000000, battery_size=6, n_top_batteries=5,forced_tasks=['rest_task'],verbose= False)
print('size 6 pfc battery:')
print(pfc_df_6['battery'][0])
pfc_df_8= bat.get_top_batteries(data_pfc,info,n_samples=1000000, battery_size=8, n_top_batteries=5,forced_tasks=['rest_task'],verbose= False)
print('size 8 pfc battery:')
print(pfc_df_8['battery'][0])
```
